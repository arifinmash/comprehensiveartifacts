```python
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv('/media/masharifin/D Drive/ComprehensiveExam/Datasets/CIC-Malmem/New/newmalwaredata80.csv')
data['Class'] = data['Class'].replace({'Benign': 0, 'Malware': 1})
X = data.drop(columns=['Class'])
y = data['Class']

desired_samples = {
    0: 29298,  # Benign
    1: 29298,  # Malware
    
}

# Initialize SMOTE
smote = SMOTE(sampling_strategy=desired_samples)

# Oversample the data using SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)

# The resampled data will be in X_resampled and y_resampled
# Now we can use this resampled data for training your model
# Convert the resampled data back to a DataFrame
resampled_data = pd.DataFrame(X_resampled, columns=X.columns)
resampled_data['Class'] = y_resampled

# Save the resampled data to a new CSV file
resampled_data.to_csv('/media/masharifin/D Drive/ComprehensiveExam/Datasets/CIC-Malmem/New/newcicresampled-smote.csv', index=False)
```
