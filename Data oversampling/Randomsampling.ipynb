```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Replace this with your actual data loading process
df= pd.read_csv('/media/masharifin/D Drive/ComprehensiveExam/Datasets/CIC-Malmem/original_test.csv')
```

```python
#hndling the NaN value
from sklearn.impute import KNNImputer

# Specify the columns with missing values
columns_to_impute = ['ldrmodules.not_in_load_avg','ldrmodules.not_in_init_avg','ldrmodules.not_in_mem_avg']

# Create a subset of the DataFrame with the selected columns
subset_df = df[columns_to_impute]

# Initialize the KNNImputer
imputer = KNNImputer(n_neighbors=2)

# Fit and transform the imputer on the selected columns
imputed_df = imputer.fit_transform(subset_df)

# Replace the original columns with the imputed values
df[columns_to_impute] = imputed_df
```

```python
nan_values = df.isna().sum()
print(nan_values)
```

```python
#df['Class'] = df['Class'].replace({'Benign': 0, 'Malware': 1})
X = df.drop(columns=['Class'])
y = df['Class']
```

```python
# Set the desired number of samples for both class 0 and class 1 (e.g., 1000 each)
desired_samples_class_0 = 29298
desired_samples_class_1 = 29298

# Initialize RandomOverSampler
ros = RandomOverSampler(sampling_strategy={0: desired_samples_class_0, 1: desired_samples_class_1}, random_state=42)
```

```python
# Apply Random Oversampling to generate exactly 29298 synthetic samples for each class
X_resampled, y_resampled = ros.fit_resample(X, y)
```

```python
# The resampled data will be in X_resampled and y_resampled
# Now you can use this resampled data for training your model
# Convert the resampled data back to a DataFrame
resampled_data = pd.DataFrame(X_resampled, columns=X.columns)
resampled_data['Class'] = y_resampled
```

```python
# Save the resampled data to a new CSV file
resampled_data.to_csv('/media/masharifin/D Drive/ComprehensiveExam/Datasets/CIC-Malmem/balanced_testdata_randomsample.csv', index=False)
```
